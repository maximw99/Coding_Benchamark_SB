```python
import pandas as pd
import json
```

```python
def read(file: str):
    res = []
    with open(file, "r") as file:
        json_obj = json.load(file)
    for j in json_obj:
        k = ("diff: ", j["diff"], "kernel: ", j["kernel"], "lex_sim: ", j["lex_sim"])
        res.append(k)
    return res
```

```python
res = read("kernel_lexsim.jsonl")
```

```python
df = pd.DataFrame(res)
df.columns = ["diff", "d", "kernel", "k", "lexsim", "l"]
```
